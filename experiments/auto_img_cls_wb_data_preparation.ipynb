```python
import wandb
PROJECT_NAME = 'toy-project-dataset2'
```

```python
_DATASET_URL = 'https://storage.googleapis.com/infuseai-auto-img-cls/datasets/noodles_v1.2.zip'

!wget $_DATASET_URL

import os
import tempfile, zipfile

basename = os.path.basename(_DATASET_URL)

with zipfile.ZipFile(basename, 'r') as zip_ref:
    zip_ref.extractall(".")

base, ext = os.path.splitext(basename)
dataset_dir = f"{base}"


print(f"dataset dir: {dataset_dir}")
```

```python
wandb.init(project=PROJECT_NAME, name='dataset', job_type='data-processing')

# path, name, type
wandb.log_artifact(dataset_dir, 'noodles', 'dataset')
wandb.finish()
```
